# feature engineering

```python
def parse_timestamp(ts):
    try:
        return pd.to_datetime(ts, format='%Y-%m-%d %H:%M:%S')
    except ValueError:
        return pd.to_datetime(ts, format='%d-%m-%Y %H:%M')


########### feature selection and doing feature engineering only on those features.#######################

def feature_engineering(df):
    new_cols = {}

    new_cols['hour'] = df.index.hour
    new_cols['day'] = df.index.day
    new_cols['month'] = df.index.month
    new_cols['day_of_week'] = df.index.dayofweek

    target_columns = ['Timestamp','AQI_calculated']

    # lag and rolling features
    for col in df.columns:
        if col not in ['hour', 'day', 'month', 'day_of_week'] + target_columns:
            for lag in [1, 3, 6, 12, 24]:
                new_cols[f'{col}_lag_{lag}'] = df[col].shift(lag)

            for window in [3, 6, 12, 24]:
                new_cols[f'{col}_rolling_{window}'] = df[col].rolling(window=window).mean()

    new_features_df = pd.concat([df, pd.DataFrame(new_cols, index=df.index)], axis=1)
    new_features_df = new_features_df.dropna()

    return new_features_df
```

# Creating Sequence

```python
def prepare_sequences(X, y, sequence_length, prediction_horizon=6):
    X_full = np.hstack((X, y))  # Stack along the last axis to combine features and AQI

    ################### try without x_full ############################
    
    X_seq, y_seq = [], []

    for i in range(len(X) - sequence_length - prediction_horizon + 1):
        X_seq.append(X_full[i : i + sequence_length])  # Past `sequence_length` timesteps + along wiht AQI
        y_seq.append(y[i + sequence_length : i + sequence_length + prediction_horizon])  # Future `prediction_horizon` values

    return np.array(X_seq), np.array(y_seq)
```

```python
# def build_model(hp, input_shape, prediction_horizons):
#     model = Sequential()
#     # First LSTM layer
#     model.add(LSTM(
#         units=hp.Int('units_1', min_value=32, max_value=512, step=32),
#         activation=hp.Choice('activation_1', values=['tanh', 'relu']),
#         return_sequences=True,
#         input_shape=input_shape
#     ))
#     model.add(Dropout(hp.Float('dropout_1', min_value=0.0, max_value=0.5, step=0.05)))
#     # Second LSTM layer
#     model.add(LSTM(
#         units=hp.Int('units_2', min_value=32, max_value=512, step=32),
#         activation=hp.Choice('activation_2', values=['tanh', 'relu']),
#         return_sequences=False
#     ))
#     model.add(Dropout(hp.Float('dropout_2', min_value=0.0, max_value=0.5, step=0.05)))
#     model.add(Dense(prediction_horizons))
#     learning_rate = hp.Float('learning_rate', min_value=1e-5, max_value=1e-2, sampling='log')
#     model.compile(optimizer=Adam(learning_rate=learning_rate), loss=hp.Choice('loss', values=['mse', 'huber_loss']))
#     return model

def build_model(hp, input_shape, prediction_horizons):
    model = Sequential()
    model.add(LSTM(units=hp.Int('units_1', min_value=32, max_value=512, step=32), return_sequences=True, input_shape=input_shape))
    model.add(Dropout(hp.Float('dropout_1', min_value=0.0, max_value=0.5, step=0.05)))
    model.add(LSTM(units=hp.Int('units_2', min_value=32, max_value=512, step=32)))
    model.add(Dropout(hp.Float('dropout_2', min_value=0.0, max_value=0.5, step=0.05)))
    model.add(Dense(prediction_horizons))
    model.compile(optimizer=Adam(hp.Float('learning_rate', min_value=1e-5, max_value=1e-1, sampling='log')), loss='mse')
    return model
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Step 1: Split data FIRST (Prevent Data Leakage)
split_row = 35041
df_train = df_feat.iloc[:split_row]  # Training set
df_test = df_feat.iloc[split_row - 24:]  # Include overlap for test sequences
```

```python
# Step 2: Initialize scalers
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

# Step 3: Fit & Transform training data, Transform test data using train scaler
X_train_raw = df_train.drop(columns=['AQI_calculated']).values
y_train_raw = df_train[['AQI_calculated']].values

X_test_raw = df_test.drop(columns=['AQI_calculated']).values
y_test_raw = df_test[['AQI_calculated']].values

X_train_scaled = scaler_X.fit_transform(X_train_raw)
X_test_scaled = scaler_X.transform(X_test_raw)  # Transform test using train scaler

y_train_scaled = scaler_y.fit_transform(y_train_raw)
y_test_scaled = scaler_y.transform(y_test_raw)

# Step 4: Prepare sequences
X_train, y_train = prepare_sequences(X_train_scaled, y_train_scaled, sequence_length=24)
X_test, y_test = prepare_sequences(X_test_scaled, y_test_scaled, sequence_length=24)

# Step 5: Define input shape for LSTM
input_shape = (X_train.shape[1], X_train.shape[2])  # (timesteps, features)
```

```python
# Hyperparameter tuning
from keras_tuner import RandomSearch
prediction_horizon = 6 

tuner = RandomSearch(
    lambda hp: build_model(hp, input_shape=input_shape, prediction_horizons=prediction_horizon),
    objective='val_loss',
    max_trials=10,
    executions_per_trial=1,
    directory='tuning_results',
    project_name='aqi_prediction'
)

early_stopping = EarlyStopping(patience=10, restore_best_weights=True)

# Searching the best hyperparameters
tuner.search(X_train, y_train, epochs=100, validation_split=0.2, callbacks=[early_stopping])

# best hyperparameters are here
best_hp = tuner.get_best_hyperparameters(num_trials=1)[0]
```
