```python
# Step 1: Replace NaN values with 0 in columns that contain NaN
columns_with_nan = df.columns[df.isnull().any()]
new_df[columns_with_nan] = new_df[columns_with_nan].fillna(0)

# Step 2: Calculate the percentage of 0s in each column
zero_percentage = (new_df == 0).mean() * 100  # Percentage of 0s in each column

# Step 3: Filter columns where the percentage of 0s is less than 30%
columns_less_than_30_percent_zeros = zero_percentage[zero_percentage < 30].index

# Extract those columns from the DataFrame
df_filtered = new_df[columns_less_than_30_percent_zeros]
```

```python
feature_col_to_impute = df_filtered.columns[1:]
feature_col_to_impute  #(these are the columns have nan value less than 30 percent)
```

```python
# Step 1: Replace 0s with NaN in the feature columns of df_with_zeros
new_df[feature_col_to_impute] = new_df[feature_col_to_impute].replace(0, np.nan)
```

```python
new_df
```

```python
from sklearn.impute import KNNImputer
```

```python
data = new_df.copy()  # Replace with your data file
```

```python
missing_values =  ((data.isnull().sum()/data.shape[0])*100)
missing_values
```

```python
param_grid = {
    'n_neighbors': [20, 22, 23, 25, 27, 28, 30, 32, 33, 35, 37, 39, 40],
    # 'n_neighbors': [7, 10, 12, 13, 15, 17, 18],
    'weights': ['uniform', 'distance']
}
```

```python
data_with_zeros = new_df.copy()
data_with_zeros = data_with_zeros.fillna(0)
```

```python
numerical_columns = data.columns[1:]
numerical_columns
```

```python
imputed_data_dict = {}
```

```python
from sklearn.metrics import mean_squared_error
# Loop over each parameter combination
for n_neighbors in param_grid['n_neighbors']:
    for weight in param_grid['weights']:
            # Create a KNN imputer with the current parameter combination
            imputer = KNNImputer(n_neighbors=n_neighbors, weights=weight)
            data_to_impute = data.copy()
            
            # Perform the imputation
            imputed_data = imputer.fit_transform(data_to_impute[numerical_columns])
            
            # Convert back to DataFrame to store it
            imputed_df = pd.DataFrame(imputed_data, columns=numerical_columns)
            
            # Save the imputed DataFrame in the dictionary
            param_key = f'n_neighbors={n_neighbors}_weights={weight}'
            imputed_data_dict[param_key] = imputed_df

            # Compare imputed data with original data (with 0 instead of NaNs)
            mse = mean_squared_error(data_with_zeros[numerical_columns], imputed_df)
            print(f"MSE for {param_key}: {mse}")
```
